```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io

data = pd.read_csv(io.BytesIO(uploaded['credit_train.csv']))
```

```python
data.head()
```

```python
data.info
```

```python
data['Purpose'].values
```

```python
# Dropping unnecessary columns:

data.drop(columns=['Loan ID', 'Customer ID', 'Months since last delinquent', 'Tax Liens', 'Years in current job'], inplace=True)
```

```python
data.head()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(data.isnull(), cmap='viridis')
```

```python
data_del = data.dropna()
```

```python
data_del.head()
```

```python
# Loan Status:

loan_status = pd.get_dummies(data_del['Loan Status'], drop_first=True)
loan_status
```

```python
data_del.head()
```

```python
# Term:

term = pd.get_dummies(data=data_del['Term'], drop_first=True)
term
```

```python
data_del.head()
```

```python
# Home Ownership

home_o = pd.get_dummies(data=data_del['Home Ownership'], drop_first=True)
home_o
```

```python
data_del.head()
```

```python
# Purpose:

purpose = pd.get_dummies(data=data_del['Purpose'], drop_first=True)
purpose
```

```python
d = data_del.drop(columns=['Loan Status', 'Term', 'Home Ownership', 'Purpose'], axis=1)
```

```python
d.head()
```

```python
de = pd.concat([d, loan_status, home_o, term, purpose], axis=1)
```

```python
de
```

```python
de.set_index('Fully Paid', inplace=True)
```

```python
de.to_csv('data_train.csv')
```
